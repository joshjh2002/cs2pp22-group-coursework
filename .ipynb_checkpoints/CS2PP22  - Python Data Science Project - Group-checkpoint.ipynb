##  1. Frame the Problem
Many people will need to apply for a loan at some point in their life. There is a lot that goes into deciding if a loan application is accepted or rejected. This project aims to fit and evaluate several different types of classification models to help see if a user may be eligible for a loan, based on their information.

This dataset contains data on previous applications for property loans and has the following attributes:
- Loan_ID: A unique loan ID.
- Gender: Either male or female.
- Married: Weather Married(yes) or Not Marttied(No).
- Dependents: Number of persons depending on the client.
- Education: Applicant Education(Graduate or Undergraduate).
- Self_Employed: Self-employed (Yes/No).
- ApplicantIncome: Applicant income.
- CoapplicantIncome: Co-applicant income.
- LoanAmount: Loan amount in thousands.
- Loan_Amount_Term: Terms of the loan in months.
- Credit_History: Credit history meets guidelines.
- Property_Area: Applicants are living either Urban, Semi-Urban or Rural.
- Loan_Status: Loan approved (Y/N).

### 5.1 Model Training

#### 5.1.1 Logistic Regression

Logistic regression is a type of supervised machine learning algorithm that can be used to model the probability of a certain class or event. It is used when the outcome variable is binary or categorical, such as yes/no, true/false, or 0/1. Logistic regression uses a function called the logistic function or the sigmoid function to transform the linear combination of the predictor variables into a probability value between 0 and 1.

```python
# Josh
# Implementation
from sklearn.linear_model import LogisticRegression

linear_regression_model, linear_regression_X_train, linear_regression_X_test, linear_regression_y_train, linear_regression_y_test = train_model(df, model=LogisticRegression(random_state=42))

logistic_regression_eval = evaluate_model(linear_regression_model, linear_regression_X_test, linear_regression_y_test, df)
```

#### 5.1.2 K-Means Clustering

K means clustering is a type of unsupervised machine learning algorithm that partitions a data set into k groups based on similarity. It works by randomly assigning each data point to one of the k clusters, then iteratively moving the data points to the cluster with the nearest mean , or centroid, and recalculating the centroids until no data points change clusters or a maximum number of iterations is reached. The goal of k means clustering is to minimize the within-cluster sum of squares, which measures the variance of the data points within each cluster.

```python
from sklearn.cluster import KMeans

kmeans_model, kmeans_X_train, kmeans_X_test, kmeans_y_train, kmeans_y_test = train_model(df, model=KMeans(n_clusters=2, random_state=42))

kmeans_eval = evaluate_model(kmeans_model, kmeans_X_test, kmeans_y_test, df)
```

```python

# Perform PCA to reduce dimensions to 2
pca = PCA(n_components=2)
pca_reduced = pca.fit_transform(kmeans_X_train)

# Plot the scatter plot
plt.scatter(pca_reduced[:, 0], pca_reduced[:, 1], c=kmeans_model.labels_, cmap='viridis')
plt.scatter(kmeans_model.cluster_centers_[:, 0], kmeans_model.cluster_centers_[:, 1], marker='x', color='red')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-Means Clustering')
plt.show()
```

#### 5.1.3 Decision Tree

A decision tree is a type of diagram that shows the possible outcomes of a series of choices. It can also be a type of machine learning algorithm that can classify or predict data based on certain rules. A decision tree has a root node, which is the starting point, and branches that lead to other nodes or leaf nodes, which are the final outcomes. A decision tree can help you make decisions by visualizing the alternatives and their consequences, or help you analyze data by finding patterns and relationships.

```python
decision_tree_model, decision_tree_X_train, decision_tree_X_test, decision_tree_y_train, decision_tree_y_test = train_model(df, model=DecisionTreeClassifier(random_state=42))

decision_tree_eval = evaluate_model(decision_tree_model, decision_tree_X_test, decision_tree_y_test, df)
```

```python
from sklearn.tree import plot_tree

# Display the decision tree
plt.figure(figsize=(12, 8))
plot_tree(decision_tree_model, feature_names=df.columns[:-1], class_names=['No', 'Yes'], filled=True)
plt.show()
```

With a 'MSE' of 0.18 and 'RMSE' of 0.42, which is average squared difference between predicted and actual values, suggests the model predictions are relativley close to the actual values, showing a strong fit to the data.
